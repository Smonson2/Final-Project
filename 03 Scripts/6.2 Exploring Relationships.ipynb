# Discussion on relationships between variables
Using world_rank as the dependent variable this graph can be used to determine the impact of the dependent variables. This is am important part of the analysis as knowing what variable(s) have the most impact on world ranking could be useful to institutions in imporving their rank by making changes with these variables. The highest correlation for world rank is (in order)rounded: 1. publications .92, 2. influence .89, 3. number of students .86, 4. patents .70, 5. quality of education .68, 6. alumni employment .67, 7. quality of faculty .67. Year had a weak relationship at .32 as did national rank at .24. Score had a negative relationship of .55. I am interested to investigate the stronger relationships and learn more of their impact on the ranking.

# Discussing the pair plot
This is a lot of information! I still want to explore the three variables mentioned earlier (publications, influence and number of students) because of the stronger relationships. I also want to take a look at year to see any changes in ranking over time. 

```python
# Creating a categorical plot
df.loc[df['world_rank'] <= 175, 'world_rank_category_loc'] = 'Low Rank'
df.loc[(df['world_rank'] >=176) & (df['world_rank'] <= 450), 'world_rank_category_loc'] = 'Low-Mid Rank'
df.loc[(df['world_rank'] >= 451) & (df['world_rank'] <= 725), 'world_rank_category_loc'] = 'Mid-High Rank'
df.loc[(df['world_rank'] >= 726) & (df['world_rank'] <= 1000), 'world_rank_category_loc'] = 'High Rank'
```

```python
df.head()
```

```python
sns.catplot(data=df, x='number_of_students', y='world_rank_category_loc')
```

```python
df['world_rank_category_loc'].value_counts(dropna = False)
```
